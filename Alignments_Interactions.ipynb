#*** Functions

```python
def compute_interaction(cmd, sRNA_seq, mRNA_seq):
  inta = !$cmd -q $sRNA_seq -t $mRNA_seq --outMode=C --seedBP=4 -n 1
  if len(inta) < 2:
    return "NA"
  else:
    return str(inta[1])
```

```python
def compute_hybrid_entries(cmd, sRNA_seq, mRNA_seq, indices):
  inta = !$cmd -q $sRNA_seq -t $mRNA_seq --seedBP=4 -n 1
  if len(inta) <= 2:
    return "NA"
  else:
    inta_hyb = [inta[i] for i in indices]
    return(inta_hyb)
```

```python
def compute_pairs(hyb1, hyb2, nuc1, nuc2):
  count = 0
  for i in range(len(hyb1)):
    if (hyb1[i] == nuc1 and hyb2[i] == nuc2) or (hyb1[i] == nuc2 and hyb2[i] == nuc1):
      count += 1
  return count
```

```python
def scatter_plot_with_stats(df, x, y):
  df = df.dropna(subset=[x, y])
  N = df.shape[0]
  graph = sns.jointplot(data=df, x=x, y=y, kind='reg', joint_kws={'line_kws':{'color':'red'}})
  plt.suptitle(f'{x} vs. {y} \n(# Interactions = {N})', y=1.05)
  r, p = stats.pearsonr(df[x], df[y])
  phantom, = graph.ax_joint.plot([], [], linestyle="", alpha=0)
  graph.ax_joint.legend([phantom], ['r={:f}, p={:e}'.format(r, p)])
  plt.show()
  print()
```
