# 2.4 Parsing the cluster to identify relevant sequences

```python
# Open the cluster text file
f = open('output70.txt.clstr', "r")
lines = f.readlines()
```

```python
# Creating a dictionary with all clusters
all_clusters = {}
cur_cluster = []
for line in lines:
  if line[0] == '>':
    if len(cur_cluster) != 0:
      cur_cluster.sort()
      all_clusters[k] = cur_cluster
    cur_cluster = []
    k = line.split('\n')[0]
  else:
    cur_cluster.append(line.split(' ')[1].split('...')[0])
cur_cluster.sort()
all_clusters[k] = cur_cluster
print("The number of all clusters are: {}".format(len(all_clusters)))
```

```python
# Creating a dictionary that holds the clusters that has all the sequences only once
all_seq_unique = {}
for k,v in all_clusters.items():
  if len(set(v)) == 20 and len(v) == 20:
    all_seq_unique[k] = v
```

```python
#Answer to 2.4a
print("The number of clusters that contains a sequence from each one of the straings in one copy only: {}".format(len(all_seq_unique)))
```

# Reverse engineering the clusters to sequences

```python
f = open('output70.txt', "r")
lines = f.readlines()
```

```python
all_actual_clusters = {}
cur_cluster = []
for line in lines:
  if line[0] == '>':
    if len(cur_cluster) != 0:
      cur_cluster.sort()
      all_actual_clusters[k] = cur_cluster
    cur_cluster = []
    k = line.split('\n')[0]
  else:
    cur_cluster.append(line)
cur_cluster.sort()
all_actual_clusters[k] = cur_cluster
print("Verifying that we parsed the same number: {}".format(len(all_clusters)))
```

```python
cluster_numbers = []
for k in all_seq_unique.keys():
  cl_number = int(k.split(' ')[1])
  cluster_numbers.append(cl_number)
print("Verifying that we collected the same amount of unique sequences: {}".format(len(cluster_numbers)))
```

```python
relevant_sequences = {}
for ii,kv in enumerate(all_actual_clusters.items()):
  if ii in cluster_numbers:
    relevant_sequences[">Cluster {}".format(ii) + kv[0].replace('>','')] = kv[1]
print(len(relevant_sequences))
```

```python
# Save the relevant sequences
with open('cluster_sequences.fasta', 'w') as f:
    for k,v in relevant_sequences.items():
        f.write(k)
        for seq in v:
          f.write(seq)
    f.close()
```

# 2.5 Adding whether a strain has CRISPR protein (Column name: "CRISPR_system")

```python
# Addming ASM codes to match the ones given by GenBank
asm_codes = []
for x in Table1["GenBank FTP"]:
  asm_codes.append("ASM"+x.split("ASM")[1])
Table1["ASM"] = asm_codes
```

```python
Table1[["ASM", "Strain"]].to_csv(path+'labels.txt', sep='\t')
```

```python
asm_gene_dict = {}
for ii in range (1,20):
  f = open(path+'2.5/strain{}_res_blastp_compact'.format(ii), "r")
  lines = f.readlines()
  if len(lines) > 0:
    for line in lines:
      split_line = line.split('\t')
      asm_gene_dict[split_line[0]] = split_line[1]
asm_gene_dict  
```

```python
crispr_genes = []
for asm in Table1["ASM"]:
  if asm in asm_gene_dict.keys():
    crispr_genes.append(asm_gene_dict[asm])
  else:
    crispr_genes.append(None)
Table1["CRISPR_system"] = crispr_genes
Table1.to_csv(path+"Table1_2.5.csv")
```
